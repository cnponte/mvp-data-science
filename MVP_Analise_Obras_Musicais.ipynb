#**Obras Musicais**

#Introdução

Este trabalho tem como objetivo analisar a distribuição de autoria em obras musicais, identificando padrões de participação entre autores e editoras.

#Dados

```python
import pandas as pd

df = pd.read_excel('Base de Dados Case Cadastro.xlsx')

df.head()
```

```python
df.columns
```

```python
df.info()
```

O dataset apresenta inconsistências estruturais, como colunas sem nome (Unnamed), indicando a necessidade de tratamento e reorganização dos dados antes da análise.

```python
df = pd.read_excel('Base de Dados Case Cadastro.xlsx', header=1)

df.head()
```

```python
df.columns
```

Foi necessário ajustar o cabeçalho do dataset, pois os nomes das colunas não estavam corretamente posicionados na primeira linha

```python
df.columns = df.columns.str.strip()

df.columns
```

```python
df = df.dropna(axis=1, how='all')

df.info()
```

Foram removidas colunas completamente vazias, que não agregavam informação à análise, contribuindo para a limpeza e organização do dataset

```python
df.head()
```

```python
df_autor1 = df[['WORK TITLE', 'AUTHOR 1', '%']].copy()

df_autor1.columns = ['WORK TITLE', 'AUTHOR', 'PERCENTUAL']

df_autor1.head()
```

Foi realizada a extração dos dados referentes ao primeiro autor de cada obra, com o objetivo de iniciar a reorganização do dataset para um formato mais adequado à análise

```python
df_autor2 = df[['WORK TITLE', 'AUTHOR 2', '%.1']].copy()

df_autor2.columns = ['WORK TITLE', 'AUTHOR', 'PERCENTUAL']

df_autor2.head()
```

```python
df_final = pd.concat([df_autor1, df_autor2])

df_final.head()
```

```python
df_final = df_final.dropna(subset=['AUTHOR'])

df_final.head()
```

```python
df_final.tail()
```

```python
df_final['PERCENTUAL'] = df_final['PERCENTUAL'] / 10000
```

```python
df_final.tail()
```

#Análise

```python
import matplotlib.pyplot as plt

top_autores = df_final['AUTHOR'].value_counts().head(10)

top_autores.plot(kind='bar')

plt.title('Top 10 autores com mais participações')
plt.xlabel('Autor')
plt.ylabel('Quantidade de obras')
plt.xticks(rotation=45)

plt.show()
```

Observa-se que alguns autores possuem maior número de participações nas obras analisadas. Isso sugere uma concentração de autoria em determinados compositores, o que pode indicar maior relevância ou recorrência desses profissionais no conjunto de dados.

```python
df_final['AUTHOR'] = df_final['AUTHOR'].replace({'Paulo Sergio Valle': 'Paulo Sérgio Kostenbader Valle'})
```

```python
df_final['AUTHOR'].value_counts().head(10)
```

```python
import unicodedata

def normalizar_nome(nome):
    if pd.isna(nome):
        return nome
    nome = nome.upper()
    nome = ''.join(
        c for c in unicodedata.normalize('NFD', nome)
        if unicodedata.category(c) != 'Mn'
    )
    return nome

df_final['AUTHOR'] = df_final['AUTHOR'].apply(normalizar_nome)
```

```python
df_final['AUTHOR'] = df_final['AUTHOR'].replace({'PAULO SERGIO VALLE': 'PAULO SERGIO KOSTENBADER VALLE'})
```

```python
df_final['AUTHOR'].value_counts().head(10)
```

Foram identificadas inconsistências na grafia dos nomes dos autores, o que poderia impactar a análise. Assim, foi realizada a padronização manual de nomes equivalentes, garantindo maior precisão nos resultados.

```python
import matplotlib.pyplot as plt

top_autores = df_final['AUTHOR'].value_counts().head(10)

top_autores.plot(kind='bar')

plt.title('Top 10 autores com mais participações')
plt.xlabel('Autor')
plt.ylabel('Quantidade de obras')
plt.xticks(rotation=45)

plt.show()
```

Agora sim, o gráfico com os 10 autores com mais participações

```python
media_autores = df_final.groupby('AUTHOR')['PERCENTUAL'].mean().sort_values(ascending=False).head(10)

media_autores.plot(kind='bar')

plt.title('Top 10 autores por média de participação (%)')
plt.xlabel('Autor')
plt.ylabel('Percentual médio')
plt.xticks(rotation=45)

plt.show()
```

A análise da média de participação por autor permite identificar aqueles que possuem maior protagonismo nas obras. Observa-se que alguns autores apresentam percentuais médios mais elevados, indicando maior participação individual nas composições.

```python
autores_por_obra = df_final.groupby('WORK TITLE')['AUTHOR'].count()

autores_por_obra.plot(kind='hist', bins=5)

plt.title('Quantidade de autores por obra')
plt.xlabel('Número de autores')
plt.ylabel('Quantidade de obras')

plt.show()
```

A análise da quantidade de autores por obra revela padrões de colaboração. Observa-se que a maioria das obras possui poucos autores, indicando que composições com grande número de colaboradores são menos frequentes

#Conclusão

A análise realizada permitiu compreender a distribuição de autoria nas obras musicais presentes no dataset. Observou-se que há uma concentração de participações em determinados autores, indicando possível centralização na produção das obras.

Além disso, a análise dos percentuais revelou padrões recorrentes de divisão de direitos autorais, como participações iguais entre dois ou mais autores.

Durante o processo, foram necessárias etapas de pré-processamento, incluindo a limpeza dos dados, reorganização da estrutura e padronização dos nomes dos autores, garantindo maior consistência e confiabilidade na análise.

Dessa forma, o MVP desenvolvido atingiu seu objetivo de explorar e gerar insights a partir dos dados disponíveis, demonstrando a importância do tratamento adequado dos dados para a obtenção de resultados relevantes.
